```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```
df=pd.read_csv('survey_results_public.csv')
df
```

```
ind = df[df['Country'] =='India']
ind
```

```
ind.dropna(inplace=True)    
ind  
```

```
plt.xlabel='Age of Developers in India'
plt.ylabel='Total'
plt.hist(ind['Age'],bins=5)
```

```
gn=pd.Series(df['Gender'].value_counts())
ey=np.array(gn.keys())
plt.pie(gn,labels=ey,autopct='%.1f%%')
plt.title("Genderwise Participation")
plt.legend(loc='lower right',bbox_to_anchor=(1.5,0.5))
plt.show()
```
