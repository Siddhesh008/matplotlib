```python
#Which  is most used  programming langauge (use LanguageWorkedWith column)
pd.DataFrame(df['LanguageWorkedWith'].value_counts())
```

```python
#Display how many work remotely on pie chart  [use WorkRemote column]
import numpy as np
import matplotlib.pyplot as plt
aa=pd.Series(df['WorkRemote'].value_counts())
k=np.array(aa.keys())
plt.pie(aa,labels=k)
plt.legend(loc='center left',bbox_to_anchor=(1.5,0.5) )
plt.title('Remotely Working Developers')
plt.axis('equal')
plt.show()
```

```python
#Display how many male and female programmers on pie chart  [use Gender column]
g=pd.Series(df['Gender'].value_counts())
ky=np.array(g.keys()) 
plt.pie(g,labels=ky)
plt.legend(loc='center left',bbox_to_anchor=(1.5,0.5))
plt.show()
```
